```
import tensorflow as tf
```

Original Multidimensional(2D) data 

```
data = [[2,1],
        [3,4],
        [6,5],
        [7,8],
        [9,10]]
```

Reshaping array


1.   Flatten
2.   Transpose



```
# reshape

x = tf.reshape(data,[-1]) # flatten the array
print(x)
x = tf.reshape(data,[2,5]) # transpose the array
print(x)
```

Aggregation

Sum , Mean etc

```
# aggregation

x = tf.reduce_max(data,-1) # max along axis
print(x)

x = tf.reduce_sum(data,-1) # sum along axis
print(x)
x = tf.reduce_sum(data) # sum all
print(x)

data_f = tf.cast(data,tf.float32) # casting to float
x = tf.reduce_mean(data_f,-1) # find mean
print(x)
```

Argmax and computation of equality

```
# argmax
x = tf.argmax(data,-1) # indices of max
print(x)

labels = [1,1,0,1,0]
corrects = tf.equal(labels,x)
print(corrects)

corrects = tf.cast(corrects,tf.float32)
accuracy = tf.reduce_mean(corrects)
print(accuracy)
```

Another multidimension array

```
second_data = [[4,2,1,3],
               [1,4,3,7]
              ]
```

Matrix multiplication and slicing of array

```
y = tf.matmul(data,second_data)
print(y)

z = tf.slice(y,[0,1],[1,3])
print(z)
```
